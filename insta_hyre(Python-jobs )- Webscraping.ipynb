```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd 
```

```python

driver = webdriver.Chrome()
driver.get("https://www.instahyre.com/python-jobs")
time.sleep(2)

#  total number of pages this could extract total 1000 job listings
total_pages = 50 

# creating a dataset to store all the data 
data = []   

# using for loop to extract data by iterating through pages 
for page in range(total_pages):
    # code to extract data from the current page by using xpath
    name_list = driver.find_elements("xpath", "//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//div[@class='employer-job-name']")
    location_list = driver.find_elements("xpath", "//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//div[@class='employer-locations']")
    founder_list = driver.find_elements("xpath", "//span[@ng-if='opp.employer.company_founded']/span[@class='ng-binding']")
    employees_list = driver.find_elements("xpath", "//span[@ng-if='opp.employer.employee_count']/span[@class='ng-binding']")
    about_list = driver.find_elements("xpath","//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//div[@class='employer-notes ng-binding ng-scope']")
    skills_list = driver.find_elements("xpath", "//div[@class='employer-details col-md-10 col-sm-8 col-xs-10']//ul[@class='tags candidate-opp-keywords']")
    link_list = driver.find_elements("xpath", "//div[@class='opportunity-go col-xs-1']/a")

    # Calculating the minimum length among all the lists to prevent  index out of range errors

    min_length = min(len(name_list), len(location_list), len(founder_list), len(employees_list), len(about_list), len(skills_list), len(link_list))



    for i in range(len(name_list)):
        name = name_list[i].text
        location = location_list[i].text
        # If the   current index 'i' is within the valid range of 'founder_list'then  assign the 'text' property of the 'founder' element to the variable, else assign an empty string
        founder = founder_list[i].text if i < len(founder_list) else '' 
        employees = employees_list[i].text
        about = about_list[i].text if i < len(about_list) else ''
        skills = skills_list[i].text.replace('\n', ', ')
        link = link_list[i].get_attribute('href')
       
       # appending the data into the list 
        data.append({
            'Name': name,
            'Location': location,
            'Founder': founder,
            'Employees': employees,
            'About': about,
            'Skills': skills,
            'Link': link
        })

    # Click the "Next" button to go to the next page
    try:
        next_button = driver.find_element("xpath", "//div[@class='pagination ng-scope']/li[@ng-click='nextPage()']")
        next_button.click()
        time.sleep(2)  # Adding sleep to allow the page to load
    except Exception :
        print("Error_clicking_next")
        break

# Creating a DataFrame from the collected data
final_df = pd.DataFrame(data)
print(final_df)

# converting to dataframe to csv file 

final_df.to_csv('python_jobs.csv', index=False)

driver.quit()
```
