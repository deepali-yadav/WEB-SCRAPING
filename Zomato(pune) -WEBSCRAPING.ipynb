```python
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.common.exceptions import WebDriverException
import pandas as pd
from urllib.parse import urljoin
import requests

# Initialize the webdriver
driver = webdriver.Chrome()
driver.get("https://www.zomato.com/pune")
time.sleep(2)

# Scroll to the end of the page
scroll_pause_time = 3
screen_height = driver.execute_script("return window.screen.height;")
i = 1

try:
    while True:
        driver.execute_script(f"window.scrollTo(0, {screen_height}*{i});")
        i += 1
        time.sleep(scroll_pause_time)
        scroll_height = driver.execute_script("return document.body.scrollHeight;")
        if (screen_height * i) > scroll_height:
            break
except WebDriverException as e:
    print(f"Exception during scrolling: {e}")
    driver.quit()
    raise

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(driver.page_source, "html.parser")
divs1 = soup.findAll('div', class_='sc-1mo3ldo-0')
divs = divs1[4:]  # Adjust based on the structure of the page

# Initialize lists to store data
urls, rest_name, ratings, address, price, cuisine = [], [], [], [], [], []

for parent in divs:
    try:
        skip_parent = False
        
        for h3_text in ["Inspiration for your first order", "Best Food in", "Top brands for", "End of search results"]:
            if parent.find("h3") and h3_text in parent.find("h3").text:
                skip_parent = True
                if h3_text == "End of search results":
                    break
                else:
                    continue

        if skip_parent:
            continue
        
        divas = parent.findAll('div', class_='jumbo-tracker')
        
        for par in divas:
            link_tag = par.find("a")
            if 'href' in link_tag.attrs:
                base = "https://www.zomato.com"
                link = link_tag.get('href')
                url = urljoin(base, link)
                urls.append(url)
                
                # Extract general details from the main page
                rating_tag = par.div.a.next_sibling.div.div.div.div.div.div.div.text
                price_tag = par.div.a.next_sibling.p.next_sibling.text
                cuisine_tag = par.div.a.next_sibling.p.text

                ratings.append(rating_tag)
                price.append(price_tag)
                cuisine.append(cuisine_tag)

                driver.get(url)
                time.sleep(3)
                soup = BeautifulSoup(driver.page_source, "html.parser")

                # Extract details from the restaurant's page
                rest_name_tag = soup.select_one('h1.sc-7kepeu-0')
                rest_name.append(rest_name_tag.text if rest_name_tag else "N/A")
                
                address_tag = soup.find('a', class_='sc-cmTdod')
                address.append(address_tag.text if address_tag else "N/A")

    except Exception as e:
        print(f"Error processing div: {e}")
        continue

# Debug print statements to check lengths of all lists
print("Length of urls:", len(urls))
print("Length of rest_name:", len(rest_name))
print("Length of addresse:", len(address))
print("Length of ratings:", len(ratings))
print("Length of price:", len(price))
print("Length of cuisine:", len(cuisine))

# Ensure all lists are of the same length
min_length = min(len(urls), len(rest_name), len(address), len(ratings), len(price), len(cuisine))
urls = urls[:min_length]
rest_name = rest_name[:min_length]
addresses = address[:min_length]
ratings = ratings[:min_length]
price = price[:min_length]
cuisine = cuisine[:min_length]

out_df = pd.DataFrame({
    'links': urls, 
    'names': rest_name, 
    'address': addresses, 
    'ratings': ratings, 
    'price for two': price,
    'cuisine': cuisine
})
out_df.to_csv("restaurants_Main.csv", index=False)

# Extracting dish data
restaurant_name = []
dish_name = []
dish_price = []

count = 0

for i in urls:
    count += 1
    print(count, end=" ")
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) AppleWebKit/537.36 (KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'
    }
    r = requests.get(i, headers=headers).text
    sou = BeautifulSoup(r, 'lxml')
    for i in range(10):
        try:
            j = sou.find_all("h4", {"class": "sc-1s0saks-15 iSmBPS"})
            dish_name.append(j[i].text)
        except:
            dish_name.append('NA')
        try:
            j = sou.find_all("span", {"class": "sc-17hyc2s-1 cCiQWA"})
            dish_price.append(j[i].text)
        except:
            dish_price.append('NA')
        try:
            j = sou.find_all("h1", {"class": "sc-7kepeu-0 sc-eilVRo eAhpQG"})
            restaurant_name.append(j[0].text)
        except:
            restaurant_name.append('NA')

# Create DataFrame for dish data
dish_df = pd.DataFrame({
    'restaurant_name': restaurant_name,
    'dish_name': dish_name,
    'dish_price': dish_price
})

# Merge main restaurant data with dish data
main_df = pd.read_csv("restaurants_Main.csv")
merged_df = pd.merge(main_df, dish_df, left_on='links', right_on='Restaurant_link')

# Save to CSV
merged_df.to_csv("zomato_1.csv", index=False)

driver.close()
```
